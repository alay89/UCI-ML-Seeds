```python
from sklearn.metrics import log_loss
def BIC(model,data,target):
    n=data.shape[0]
    p=data.shape[1]
    complexity=np.log(n)*(p+1)
    model.fit(data,target)
    pred=model.predict_proba(data)
    likelihood=-2*log_loss(target,pred)
    return complexity - likelihood 
```

```python
BIC(result2, X2, y) #model with 3 features from RFE selection
```

```python
BIC(result, X, y) #model with all features
```
